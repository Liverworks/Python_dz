```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ib/python/14.pandas/who_suicide_statistics.csv')
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df = df.dropna(axis=0) # still enough data
```

```python
df['suicides_no'].corr(df['population'])
```

```python
df['suicides_no'].corr(df['year'])
```

```python
df[['country', 'suicides_no']].groupby(['country']).agg('mean').sort_values(by = 'suicides_no', ascending = False).head(10)
```

```python
df[['year', 'suicides_no']].groupby(['year']).agg('sum').sort_values(by = 'suicides_no', ascending = False).head(10)
```

```python
plt.rcParams['figure.figsize'] = (18, 9)

x = pd.DataFrame(df.groupby(['year'])['suicides_no'].sum().reset_index())
x.sort_values(by = ['suicides_no'], ascending = False, inplace = True)

sns.lineplot(x['year'], y = x['suicides_no'], data = x, palette = 'cool')
plt.title('Distribution of suicides from the year 1985 to 2016', fontsize = 20)
plt.xlabel('year')
plt.xticks(rotation = 90)
plt.ylabel('count')
plt.show()
```

```python
df2 = df.groupby(['country'])['suicides_no', 'population'].mean()
df2 = pd.DataFrame(df2)
sns.scatterplot(x='population', y='suicides_no', data=df2)
plt.title('Mean suicides vs counry population', fontsize = 20)
plt.grid()
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (18, 7)


sns.stripplot(df['age'], df['suicides_no'], palette = 'Wistia')
plt.title('Age groups vs Sexes', fontsize = 20)
plt.grid()
plt.show()
```

```python
sns.boxenplot(df['sex'], y = df['suicides_no'], data = x)
plt.title('Distribution of suicides of sexes', fontsize = 20)
plt.xlabel('year')
plt.xticks(rotation = 90)
plt.ylabel('count')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (18, 7)

df2 = df.groupby(['country', 'year'])['suicides_no'].mean()
df2 = pd.DataFrame(df2)
df2 = df2.reset_index(level=['year'])
sns.stripplot(df2['year'], df2['suicides_no'])
plt.title('Year vs Suicides', fontsize = 20)
plt.xticks(rotation = 90)
plt.show()
```

```python
df['suicide_percent'] = df.suicides_no / df.population * 100  # but in percents, not in absolute numbers, the picture is different
```

```python
df[['country', 'suicide_percent']].groupby(['country']).agg('mean').sort_values(by = 'suicide_percent', ascending = False).head(10)
```

```python
df[['year', 'suicide_percent']].groupby(['year']).agg('sum').sort_values(by = 'suicide_percent', ascending = False).head(10)
```

```python
plt.rcParams['figure.figsize'] = (18, 9)

x = pd.DataFrame(df.groupby(['year'])['suicide_percent'].sum().reset_index())
x.sort_values(by = ['suicide_percent'], ascending = False, inplace = True)

sns.lineplot(x['year'], y = x['suicide_percent'], data = x, palette = 'cool')
plt.title('Distribution of suicide percent from the year 1985 to 2016', fontsize = 20)
plt.xlabel('year')
plt.xticks(rotation = 90)
plt.ylabel('count')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (18, 7)

df2 = df.groupby(['country', 'year'])['suicide_percent'].mean()
df2 = pd.DataFrame(df2)
df2 = df2.reset_index(level=['year'])
sns.stripplot(df2['year'], df2['suicide_percent'])
plt.title('Year vs Suicide percent', fontsize = 20)
plt.xticks(rotation = 90)
plt.show()
```
